## Regresion Algorithm

### Gradient Descent

```python
def NormalEquation(X, y):
    """
    Normal Equation untuk Linear Regression: theta = (X^T X)^-1 X^T y
    
    Input:
      X : shape (m, n) - data training (sudah include intercept column jika diperlukan)
      y : shape (m,) - target
    Output:
      theta : shape (n,) - parameter
    """
    X = np.asarray(X, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64).ravel()
    
    X = np.c_[np.ones((X.shape[0], 1)), X]
    # Normal Equation: theta = (X^T X)^-1 X^T y
    theta, _, _, _ = np.linalg.lstsq(X, y, rcond=None)
    
    return theta
```

### Regression

```python
def Regression(X, theta):
    X = np.asarray(X)
    theta = np.asarray(theta)
    X = np.c_[np.ones((X.shape[0], 1)), X]

    return X.dot(theta)
```

## Data Scaling

```python
class ZscoreNormal:
    def __init__(self):
        self.mean = None
        self.std = None
    def fit(self, X):
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
    def transform(self, X):
        return (X - self.mean) / self.std
    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)
    def inverse_transform(self, X_norm):
        return (X_norm * self.std) + self.mean
```

### train and test

```python
theta = NormalEquation(X_train, y_train)
y_pred = Regression(X_test, theta)
```

## evaluation

```python
def MSE(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
def RMSE(y_true, y_pred):
    return np.sqrt(MSE(y_true, y_pred))
def r2_score(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    return 1 - (ss_residual / ss_total)
from sklearn.metrics import mean_absolute_percentage_error

```

```python
mse = MSE(y_test, y_pred)
rmse = RMSE(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
print("MAPE:", mape)
```

```python
import numpy as np
import pandas as pd

# ==============================================================================
# FUNGSI GENERATOR TABEL LATEX OTOMATIS
# ==============================================================================
# Pastikan Anda sudah memiliki variabel ini di notebook:
# 1. df      : Dataframe asli
# 2. y_test  : Target asli (numpy array atau list)
# 3. y_pred  : Hasil prediksi (numpy array atau list)
# ==============================================================================

def print_latex_tables(df, y_test, y_pred):
    
    # --- TABEL 1: SAMPEL DATASET (5 Baris Pertama) ---
    print("% ==========================================")
    print("% TABEL 1: FORMAT BOOKTABS (COPY KE LATEX)")
    print("% ==========================================")
    print("\\begin{table}[htbp]")
    print("    \\caption{Sampel Dataset Harga Rumah (5 Baris Pertama)}")
    print("    \\begin{center}")
    print("    \\begin{small}")
    print("    \\begin{tabular}{cccccc}")
    print("        \\toprule")
    print("        LT ($m^2$) & LB ($m^2$) & K.Tidur & K.Mandi & Garasi & Harga (M) \\\\")
    print("        \\midrule")
    
    # Mengambil 5 data teratas dari df asli
    for i in range(5):
        row = df.iloc[i]
        # Mengubah harga ke Miliar (dibagi 1 Miliar)
        harga_m = row['harga_rumah'] / 1_000_000_000
        print(f"        {int(row['luas_tanah'])} & {int(row['luas_bangunan'])} & "
              f"{int(row['jumlah_kamar_tidur'])} & {int(row['jumlah_kamar_mandi'])} & "
              f"{int(row['garasi'])} & {harga_m:.1f} \\\\")
              
    print("        \\bottomrule")
    print("    \\end{tabular}")
    print("    \\end{small}")
    print("    \\label{tab:rawdata}")
    print("    \\end{center}")
    print("\\end{table}")
    print("\n")


    # --- TABEL 2: METRIK EVALUASI ---
    # Hitung ulang metrik agar akurat sesuai data saat ini
    mse = np.mean((y_test - y_pred) ** 2)
    rmse = np.sqrt(mse)
    ss_res = np.sum((y_test - y_pred) ** 2)
    ss_tot = np.sum((y_test - np.mean(y_test)) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    print("% ==========================================")
    print("% TABEL 2: HASIL EVALUASI")
    print("% ==========================================")
    print("\\begin{table}[htbp]")
    print("    \\caption{Metrik Evaluasi Kinerja Model}")
    print("    \\begin{center}")
    print("    \\begin{small}")
    print("    \\begin{tabular}{lc}")
    print("        \\toprule")
    print("        Metrik & Nilai \\\\")
    print("        \\midrule")
    print(f"        R-squared ($R^2$) & {r2:.3f} \\\\")
    print(f"        RMSE (Miliar Rupiah) & {rmse/1e9:.2f} \\\\")
    print(f"        MAPE (Persentase Error) & {mape:.2f}\\% \\\\")
    print("        \\bottomrule")
    print("    \\end{tabular}")
    print("    \\end{small}")
    print("    \\label{tab:evaluasi}")
    print("    \\end{center}")
    print("\\end{table}")
    print("\n")


    # --- TABEL 3: PERBANDINGAN PREDIKSI (SAMPEL ACAK) ---
    print("% ==========================================")
    print("% TABEL 3: SAMPEL PREDIKSI")
    print("% ==========================================")
    print("\\begin{table}[htbp]")
    print("    \\caption{Sampel Perbandingan Harga Asli vs Prediksi Model}")
    print("    \\begin{center}")
    print("    \\begin{small}")
    print("    \\begin{tabular}{cccc}")
    print("        \\toprule")
    print("        Luas ($m^2$)* & Aktual (M) & Prediksi (M) & Selisih (M) \\\\")
    print("        \\midrule")

    # Ambil 5 indeks acak dari data test
    indices = np.random.choice(len(y_test), 5, replace=False)
    
    for idx in indices:
        act = y_test[idx]
        pred = y_pred[idx]
        diff = pred - act
        
        # Trik: Mencari Luas Tanah asli dari dataframe yang harganya cocok
        # (Karena X_test sudah diskalakan/di-transform, kita ambil dari df asli untuk display)
        # Kita cari rumah di df yang harganya paling mendekati harga aktual ini
        closest_idx = (df['harga_rumah'] - act).abs().idxmin()
        luas_tanah_asli = df.loc[closest_idx, 'luas_tanah']
        
        tanda = "+" if diff > 0 else ""
        
        print(f"        {int(luas_tanah_asli)} & {act/1e9:.1f} & {pred/1e9:.1f} & {tanda}{diff/1e9:.1f} \\\\")

    print("        \\bottomrule")
    print("    \\end{tabular}")
    print("    \\end{small}")
    print("    \\label{tab:sampel}")
    print("    \\end{center}")
    print("    {\\footnotesize *Luas Tanah diambil dari data asli yang relevan.}")
    print("\\end{table}")

# --- EKSEKUSI ---
print("Copy output di bawah ini ke file paper.tex Anda:\n")
print("-" * 60)
print("\n")
print_latex_tables(df, y_test, y_pred)
```
